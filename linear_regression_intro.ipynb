# Intro to Linear Regression - A fundamental approach

Linear Regression in ML works by gradient descent, iteractively adjusting model weights and biases until the model converges. In this tutorial, we will take a simple approach by greating a function for calculating cost and adjusting model weights. 

```python
#import as required
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#import a dataset
cars = pd.read_json('datasets/carsData.json')
cars.head()
```

For this excercise, out _label_ will be Miles_per_Gallon, and we will choose one _feature_ to that can be used to predict it. Lets start by visualising the relationship each variable has with out _label_. 

```python
# iterate over all labels in the datafram

for column in cars.columns[2:]:
    plt.scatter(cars[column], cars['Miles_per_Gallon'])
    plt.xlabel(column)
    plt.ylabel('Miles_per_Gallon')
    plt.show()
```
